#Método de Garlekin

Este es un tipo de método de residuos pesados, donde las funciones de peso

coinciden con las funciones base βᵢ


galerkin.png

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Variable simbólica
x = sp.symbols('x')

# Funciones base φ_i(x) = sin(i π x)
phi = [sp.sin(sp.pi * (i+1) * x) for i in range(3)]

# Coeficientes α_i
a1, a2, a3 = sp.symbols('a1 a2 a3')
alpha = [a1, a2, a3]
```

Solución aproximada

```python
# Aproximación u3(x)
u3 = sum(alpha[i] * phi[i] for i in range(3))

# Ecuación residual: -u'' + u - x
u3_2d = sp.diff(u3, x, 2)
R = -u3_2d + u3 - x
u3_2d, R
```

```python
# Ecuaciones de Galerkin: (R, φ_j) = 0
eqs = [sp.integrate(R * phi[j], (x, 0, 1)) for j in range(3)]
eqs
```

```python
# Resolver para a1, a2, a3
sol_alpha = sp.solve(eqs, [a1, a2, a3])
sol_alpha
```

Solución aproximada

```python
u3_approx = u3.subs(sol_alpha)
u3_approx_simplified = sp.simplify(u3_approx)
u3_approx_simplified
```

```python
u3_fn = sp.lambdify(x, u3_approx_simplified, "numpy")
```

Solución exacta

```python
u = sp.Function('u')
ode = -u(x).diff(x,2) + u(x) - x
sol_exact = sp.dsolve(ode, ics={u(0):0, u(1):0})
u_exact_fn = sp.lambdify(x, sol_exact.rhs, "numpy")
sol_exact
```

Graficamos

```python
xx = np.linspace(0, 1, 200)
plt.figure(figsize=(9,5))
plt.plot(xx, u3_fn(xx), label="Galerkin (3 términos)", linewidth=3)
plt.plot(xx, u_exact_fn(xx), "--", label="Solución Exacta", linewidth=2)

plt.title("Método de Galerkin — EDO:  -u'' + u - x = 0")
plt.xlabel("x")
plt.ylabel("u(x)")
plt.grid(True)
plt.legend()
plt.show()
```
